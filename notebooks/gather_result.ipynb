```python
from pathlib import Path
import json, re
import pandas as pd
from collections import defaultdict
```

```python
def jload(input_file):
    with open(input_file) as f:
        return json.load(f)
```

```python
input_dir = "/home/bhuang/icd_10/outputs/eval/quaero"
# input_dir = "/home/bhuang/icd_10/outputs/eval/synthetic"

results = defaultdict(list)

for p in Path(input_dir).rglob("*pred-eval.json"):
    if "valid" in p.as_posix():
        continue

    model_name = p.parent.stem
    data_split_name = re.sub(r"-pred-eval$", "", p.stem)

    r = jload(p.as_posix())
    for k, v in r.items():
        results[model_name].append((f"{data_split_name}_{k}", v))
```

```python
new_results = []
for name, metrics in results.items():
    r = {"model_name": name}
    for k, v in metrics:
        r[k] = v
    new_results.append(r)

df = pd.DataFrame(new_results)
df
```

```python
df.to_csv("tmp_result/tmp.csv")
```
